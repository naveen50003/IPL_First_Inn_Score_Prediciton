```python
#importing essentials libraries
import pandas as pd
import pickle
```

```python
# Loading dataset
df = pd.read_csv('ipl.csv')
```

```python
df.head()
```

```python
#Data cleaning
columns_to_remove = ['mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels=columns_to_remove, axis=1, inplace=True)
```

```python
df.head()
```

```python
df['bat_team'].unique()
```

```python
# keeping only consistent Teams
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
df.head()
```

```python
df['bat_team'].isin(consistent_teams)
```

```python
# removing the first 5 overs data in every match
df=df[(df['overs'])>=5.0]
```

```python
df.head()
```

```python
#converting the column 'date' from string into datetime object
from datetime import datetime
```

```python
df['date']=df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df.head()
```

```python
#-----Data Preprocessing--#
# Converting categorical features using oneHotEncoding
encoded_df =pd.get_dummies(df, columns=['bat_team','bowl_team'])
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
encoded_df = encoded_df[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5',
       'total']]
```

```python
#Splitting data in training and test set
X_train = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year <= 2016]
X_test = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year >= 2017]
```

```python
y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values
```

```python
X_train
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
corrmat = encoded_df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g = sns.heatmap(encoded_df[top_corr_features].corr(), annot=True, cmap="YlGnBu")
```

```python
X_train.drop(labels='date', axis = True, inplace=True)
X_test.drop(labels='date', axis = True, inplace=True)
```

```python
# Model Building
# Linear Regression

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Creating a pickle file for the classifier
filename='first-innings-score-lr-model.pkl';
pickle.dump(regressor, open(filename, 'wb'))
```
